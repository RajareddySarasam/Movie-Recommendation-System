```python
import ast # is to convert the string to list
# "[1, 2, 3, 4]" -> ast.literal_eval() -> [1,2,3,4]
```

```python
def preprocessing(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
# I just wanna Top 3 character.And their names are also in values of key 'name'.
def preprocessing1(text):
    counter=0
    l=[]
    for i in ast.literal_eval(text):
        if counter <3 :
            l.append(i['name'])
            counter+=1
        else:
            break
    return l  

```

```python
# It Contains about every crew member of the movie.But i just want Director of the movie.
def preprocessing2(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies.crew.apply(preprocessing2)
```

```python
# It is to remove the spaces between names (Different tokens for same name eg: 'johnny Depp' -> 'johnny','Depp')

def collapse(L):
    L1 = []
    for i in L:
        #Removing spaces between words.
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
# Removing the stopwords
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
```

```python
def stopword_removal(text):
    word_tokens = word_tokenize(text)
    
    # converting the words in word_tokens to lower case and then checking whether they are present in stop_words or not
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]   
    
    return ' '.join(filtered_sentence)
```

```python
# Stemming
```

```python
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer() 
```

```python
def stemming(text):
    word_tokens = word_tokenize(text)
    # applying stemming to each word.
    stemmed_words = [porter_stemmer.stem(word) for word in word_tokens]
    return ' '.join(stemmed_words)
```

```python
# Applying Bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['summary']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
# So,Now we have 5000 dimensions for each of the vector.
#Using the cosine similarity metric bcz euclidean distance functions improperly for higher dimensions.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    # Getting index of the movie
    index = new_df[new_df['title'] == movie].index[0]
    
    # Sorting the Distances of the vectors  
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
    # Getting Top 5 Similar Movies
    
    for i in distances[1:6]:
        # i[0] -> Gives Index
        print(new_df.iloc[i[0]].title)
```
